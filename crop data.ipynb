```python
import numpy as np
n=int(input("enter the number of months"))
for i in range(n):
  data=float(input("enter rainfall(mm):"))
  rainfall_data=[]
  rainfall_data.append(data)
  rainfall=np.array([rainfall_data])
for i in range(n):
  yield_data=float(input("enter yield(tons/ha):"))
  yields=[]
  yields.append(yield_data)
  yields_data=np.array([yield_data])
  print('===RAINFALL STATISTICS===')
print(f'mean:{np.mean(rainfall):2f}mm')
print(f'median:{np.median(rainfall):2f}mm')
print(f'std dev:{np.std(rainfall):2f}mm')
print(f'variance:{np.var(rainfall):2f}mm')
print(f'min:{np.min(rainfall)}mm(month{np.argmin(rainfall)+1})')
print(f'max:{np.max(rainfall)}mm(month{np.argmax(rainfall)+1})')
print(f'range:{np.ptp(rainfall)}mm')
print("\n===YIELD STATISTICS===")
print(f'mean:{np.mean(yields):2f}tons/ha')
print(f'median:{np.median(yields):2f}tons/ha')
print(f'std dev:{np.std(yields):2f}tons/ha')
print('\n===PERCENTILES===')
percentiles=[25,50,75,90]
rainfall_percentiles=np.percentile(rainfall,percentiles)
for p,val in zip(percentiles,rainfall_percentiles):
  print(f'{p}th percentile rainfall:{val:2f}mm')
print('\n===CORRELATION ANALYSIS===')
correlation_matrix=np.corrcoef(rainfall,yields)
correlation=correlation_matrix[0,1]
print(f'rainfall-yield correlation:{correlation:4f}')
print('interpretation:strong positive correlation' if correlation>0.7 else 'moderate correlation' if correlation>0.4 else 'weak correlation')
print('\n===CUMULATIVE RAINFALL===')
cumulative=np.cumsum(rainfall)
print('month\trainfall\tcumulative')
for month,(rain,cum)in enumerate(zip(rainfall,cumulative),1):
  print(f'{month}\t{rain}\t\t{cum}')




```
