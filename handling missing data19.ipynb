```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Name':['A','B','C','D'],
    'Age':[23,np.nan,21,np.nan],
    'Marks':[85,90,np.nan,88]
    })
print("missing values in each column:")
print(df.isna().sum())
```

```python
print("\nRows with missing values:")
print(df[df.isna().any(axis=1)])
```

```python
print("\nAfter filling missing values:")
print(df.isna().sum())


```

```python
print("\nfinal dataframe")
print(df)
```
