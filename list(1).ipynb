```python
#index()method
thislist=["apple","banana","cherry","banana"]
index_value=thislist.index("banana")
print("index of 'banana':",index_value)
```

```python
#count()method
thislist=["apple","banana","apple","cherry","apple"]
count_value=thislist.count("apple")
print("count of 'apple'",count_value)
```

```python
#reverse()method
#thislist=["orange","mango","kiwi","banana"]
print("original list:",thislist)
thislist.reverse()
print("reverse list:",thislist)
```

```python
original_list=[1,2,3,4,5]
reverse_iterator=reversed(original_list)
reversed_list=list(reverse_iterator)#convert the iterator to a list
print(reversed_list)
```
