```python
def load_data(filename):
  "read crop data from file and return a list of dictionary"
  crops = [] # Initialize as an empty list to be populated from the file
  with open(filename, "r") as file:
    for line in file:
      line = line.strip()
      if line:
        data = line.split(",")
        crop_data = {
            "name": data[0],
            "season": data[1],
            "area": float(data[2]),
            "yield": float(data[3]),
            "cost": float(data[4]),
            "revenue": float(data[5])
        }
        crops.append(crop_data)
        print(crop_data)
  return crops
```

```python
def write_report(filename,crops,season):
  """write analysis report to output file"""
  file=open(filename,"w")
  file.write("agricultural data analysis report\n\n")
  file.write("season of the crop:\n")
  for crop in crops:

    file.write(f"{crop['name']}:{crop['season']}\n")
```
