```python
crop_price={
    "rice":40,
    "wheat":35,
    "maize":30
}
crop_yield={
    "rice":1200,
    "wheat":1500,
    "maize":1100
}
for a,b in zip(crop_price,crop_yield):
  print(a,b)
```

```python
for crop,price in zip(crop_price.keys(),crop_price.values()):
 print(crop,price)
```

```python
for crop,price,yield_kg in zip(
    crop_price.keys(),
    crop_price.values(),
    crop_yield.values()):
    print(f"{crop}: price={price}, yield={yield_kg}")
```
