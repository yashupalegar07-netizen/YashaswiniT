```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()

```

```python
x=["APPLES","BANANAS"]
y=[400,350]
plt.bar(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="orange")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="violet")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width= 0.1)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,height=0.1)
plt.show()
```

```python
x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
data=np.random.normal(170,10,250)
plt.hist(data,bins=20)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram example")
plt.show()
```
