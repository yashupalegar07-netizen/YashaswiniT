```python
def increase_yield(yields,increment):
    for i in range(len(yields)):
      yields[i]+=increment
crop_yields=[1200,1350,1100]
increase_yield(crop_yields,100)
print(crop_yields)
```

```python
def convert_to_quintals(yields):
  result=[]
  for value in yields:
    result.append(value/100)
  return result
crop_yields=[1200,1350,1100]
quintals=convert_to_quintals(crop_yields)
print(quintals)
```

```python
def high_yield_crops(yields,threshold):
   high=[]
   for value in yields:
      if value>=threshold:
        high.append(value)
   return high
crop_yields=[1200,900,1500,800]
print(high_yield_crops(crop_yields,1000))
```
