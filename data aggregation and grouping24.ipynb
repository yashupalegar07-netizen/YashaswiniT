```python
import pandas as pd
data={
    'Department':['HR','Sales','IT','Sales','IT','HR','Sales'],
    'Employee':['A','B','C','D','E','F','G'],
    'Salary':[50000,60000,45000,48000,70000,72000,71000],
    'Experience':[2,4,3,5,6,7,5]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Department','Experience']).sum()
```

```python
df.groupby(['Department','Experience'])
```

```python
df3=df.groupby(['Department','Experience']).sum()
print(df3)
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df1=df.groupby(['Department','Experience']).sum()
print(df1)
```

```python
df2=df.groupby('Department').sum()
print(df2)
```

```python
df5=df.groupby('Department')['Salary'].mean()
print(df5)
```

```python
df6=df.groupby('Department')['Employee'].count()
print(df6)
```

```python
df7=df.groupby('Department')['Salary'].agg(['sum','mean','std'])
print(df7)
```

```python
df.groupby('Department').agg({
    'Salary':'sum','Experience':'mean'
})
```

```python
df9=df['Dept_Avg_Salary']=df.groupby('Department')['Salary'].transform('mean')
print(df9)
```

```python
df1=df.groupby('Department').filter(lambda x:len(x)>2)
print(df1)
```

```python
df.pivot_table(
    values='Salary',
    index='Department',
    columns='Experience',
    aggfunc='mean'

)
```

```python
import pandas as pd
data={
    'Crop':['rice','wheat','ragi','rice','ragi','wheat','rice'],
    'season':['kharif','rabi','kharif','rabi','rabi','kharif','kharif'],
    'fertilizer':[100,120,150,160,200,220,180],
    'yield':[25,28,31,27,32,29,26]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Crop','yield']).sum()
```

```python
df3=df.groupby('fertilizer').sum()
print(df3)
```

```python
df.groupby('Crop')['fertilizer'].sum()
```

```python
df1=df.groupby(['Crop','season']).sum()
print(df1)
```

```python
df2=df.groupby('Crop').sum()
print(df2)
```

```python
df5=df.groupby('Crop')['yield'].mean()
print(df5)
```

```python
df6=df.groupby('Crop')['yield'].count()
print(df6)
```

```python
df7=df.groupby('Crop')['yield'].agg(['sum','mean','std'])
print(df7)
```

```python
df.groupby('Crop').agg({
    'fertilizer': 'sum',
    'yield': 'mean'
})
```

```python
df9=df['Crop_Avg_Yield']=df.groupby('Crop')['yield'].transform('mean')
print(df9)
```

```python
df1=df.groupby('Crop').filter(lambda x:len(x)>2)
print(df1)
```

```python
df.pivot_table(
    values='yield',
    index='Crop',
    columns='fertilizer',
    aggfunc='mean'
)
```
