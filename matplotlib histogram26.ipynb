```python
import matplotlib.pyplot as plt
data=[
    30,35,40,42,45,47,48,49,50,50,51,52,53,55,57,60,62,65,70,75
]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="orange",
    edgecolor="black",
    alpha=0.5,density=True,
    range=(30,80)
)
plt.xlabel("value")
plt.ylabel("density")
plt.title("histogram without random data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
