```python
def load_data(filename):
  crops=[]
  file=open(filename,"r")
  for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
        "name":data[0],
        "season":data[1],
        "area":float(data[2]),
        "rainfall":float(data[3]),
        "fertilizer":float(data[4]),
        "yield":float(data[5])
      }
      crops.append(crop)
  file.close()
  return crops
def calculated_yield_per_area(crop):
  """return yield per unit area"""
  return crop["yield"]/crop["area"]
def display_crops(crops):
  """display all crop records"""
  print("\ncrop records:")
  for crop in crops:
    ypa=calculated_yield_per_area(crop)
    print(crop["name"],"_",crop["season"],"-yield/area:",round(ypa,2))
def find_highest_and_lowest_yield(crops):
  """finds crops with highest and lowest yield per area"""
  if not crops:
    return None, None
  highest=crops[0]
  lowest=crops[0]
  for crop in crops:
    if calculated_yield_per_area(crop)>calculated_yield_per_area(highest):
      highest=crop
    if calculated_yield_per_area(crop)<calculated_yield_per_area(lowest):
      lowest=crop
  return highest,lowest
def seasonal_summary(crops):
  """generates seasonal yield summary using dictionary"""
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculated_yield_per_area(crop)
    if season not in summary:
      summary[season]=[]
    summary[season].append(ypa)
  return summary
def write_report(filename,crops,highest,lowest,summary):
  """writes analysis result to output file"""
  file=open(filename,"w")
  file.write("agricultural data analysis report\n\n")
  file.write("crop yield per unit area:\n")
  for crop in crops:
    ypa=calculated_yield_per_area(crop)
    file.write(crop["name"]+":"+str(round(ypa,2)) + "\n")

  if highest:
    file.write("\nHighest yield crop:\n")
    file.write(highest["name"] + "\n")

  if lowest:
    file.write("Lowest yield crop:\n")
    file.write(lowest["name"] + "\n")

  file.write("\nseasonal average yield:\n")
  for season in summary:
    if summary[season]:
      avg=sum(summary[season])/len(summary[season])
      file.write(season+":"+str(round(avg,2)) + "\n")
  file.close()

# Define crop_data_string using the content of the uploaded file.
crop_data_string = upload['Document agri (2).txt'].decode('utf-8')

# Preprocess crop_data_string to fix malformed lines
crop_data_string = crop_data_string.replace('1800 Soybean', '1800\nSoybean')
crop_data_string = crop_data_string.replace('1400 Sunflower', '1400\nSunflower')
crop_data_string = crop_data_string.replace('1700 Potato', '1700\nPotato')

#----------main program----------
def main():
  input_file="crops.txt"
  output_file="analysis_report.txt"

  # Write the crop data to crops.txt before loading it
  with open(input_file, "w") as f:
      f.write(crop_data_string.strip())

  crops=load_data(input_file)

  if not crops:
    print(f"No data loaded from {input_file}. Exiting.")
    return

  for crop in crops:
    print(crop)
  for crop in crops:
    for key,value in crop.items():
      print(f"{key}:{value}")
      print("-----")
  display_crops(crops)
  highest,lowest=find_highest_and_lowest_yield(crops)
  summary=seasonal_summary(crops)
  write_report(output_file,crops,highest,lowest,summary)
  print("\nanlysis is complete report written to",output_file)

main()
```

```python
def crop_count_per_season(crops):
  count={}
  for crop in crops:
    season=crop["season"]
    count[season] = count.get(season, 0) + 1
  for season, value in count.items():
    print(season,":",value)
input_file="crops.txt"
crops=load_data(input_file)
crop_count_per_season(crops)
```

```python
def fertilizer_efficiency(crop):
  return crop["yield"]/crop["fertilizer"]
def most_fertilizer_efficient_crop(crops):
  best=crops[0]
  for crop in crops:
    if fertilizer_efficiency(crop)>fertilizer_efficiency(best):
      best=crop
  print("most fertilizer efficient crop:",best["name"])
input_file="crops.txt"
crops=load_data(input_file)
most_fertilizer_efficient_crop(crops)
```

```python
def rainfall_efficiency_analysis(crops,max_rainfall):
  print("crops with ralinfall<=",max_rainfall)
  for crop in crops:
    if crop["rainfall"]<=max_rainfall:
      print(crop["name"],"-",crop["yield"])
input_file="crops.txt"
crops=load_data(input_file)
max_rainfall_value = float(input("Enter maximum rainfall for analysis: "))
rainfall_efficiency_analysis(crops, max_rainfall_value)
```

```python
def rank_crops_by_yield(crops):
  ranked=sorted(crops,key=calculated_yield_per_area,reverse=True)
  rank=1
  for crop in ranked:
    print(rank,crop["name"],"-",round(calculated_yield_per_area(crop),2))
    rank+=1
input_file="crops.txt"
crops=load_data(input_file)
rank_crops_by_yield(crops)
```

```python
def best_crop_per_season(crops):
  best={}
  for crop in crops:
    season=crop["season"]
    if season not in best:
      best[season]=crop
    else:
      if calculated_yield_per_area(crop)>calculated_yield_per_area(best[season]):
        best[season]=crop
  for season,crop in best.items():
    print(season,":",crop["name"])
input_file="crops.txt"
crops=load_data(input_file)
best_crop_per_season(crops)
```

```python
def crops_in_multiple_seasons(crops):
  crop_seasons={}
  for crop in crops:
    name=crop["name"]
    season=crop["season"]
    if name not in crop_seasons:
      crop_seasons[name]=[]
    if season not in crop_seasons[name]:
      crop_seasons[name].append(season)
  for name,seasons in crop_seasons.items():
    if len(seasons)>1:
      print(name,"grown in seasons:",seasons)
input_file="crops.txt"
crops=load_data(input_file)
crops_in_multiple_seasons(crops)
```

```python
def recommended_crops(crops,season,max_rainfall,max_fertilizer):
  print("recommended crops:")
  for crop in crops:
    if(crop["season"]==season and
       crop["rainfall"]<=max_rainfall and
       crop["fertilizer"]<=max_fertilizer):
      print(crop["name"])
input_file="crops.txt"
crops=load_data(input_file)

season_input = input("Enter the season for recommendation (e.g., Kharif, Rabi, Annual): ")
max_rainfall_input = float(input("Enter maximum rainfall for recommended crops: "))
max_fertilizer_input = float(input("Enter maximum fertilizer for recommended crops: "))

recommended_crops(crops, season_input, max_rainfall_input, max_fertilizer_input)
```
