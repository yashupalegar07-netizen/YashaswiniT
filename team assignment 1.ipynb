```python
def crop_data():
  crops=[]
  file=open("agri data.txt","r")
  for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop_data={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "revenue":float(data[5])
    }
    crops.append(crop_data)
  file.close()
  return crops
crops=crop_data()
print(crops)
```

```python
print("kharif crops:")
for crop in crops:
  if crop["season"]=="kharif":
    print(crop["name"])
```

```python
print("rabi crops:")
for crop in crops:
  if crop["season"]=="rabi":
    print(crop["name"])

```

```python
print("non kharif crops:")
for crop in crops:
  if crop["season"]!="non kharif":
    print(crop["name"])

```

```python
def calculate_profit(crop):
  return crop["revenue"]-crop["cost"]

print("profit for each crop:")
for crop in crops:
  profit=calculate_profit(crop)
  print(crop["name"],":",profit)
```

```python
total_area=0
total_yield=0
for crop in crops:
  total_area+=crop["area"]
  total_yield+=crop["yield"]
average_yield=total_yield/len(crops)
print("summary statistics:")
print("total area:",total_area)
print("average yield:",average_yield)
```
